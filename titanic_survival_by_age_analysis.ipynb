```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset
titanic = sns.load_dataset("titanic")

# Check missing age values
missing_ages = titanic["age"].isnull().sum()
print(f"Number of missing age values: {missing_ages}")

# Keep only rows where age is available
titanic_age = titanic.dropna(subset=["age"])

# Histogram by survival
plt.figure(figsize=(10, 6))
sns.histplot(
    data=titanic_age,
    x="age",
    bins=30,
    kde=True,
    hue="survived",
    multiple="stack"
)
plt.title("Age Distribution by Survival")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

# KDE plot by survival
plt.figure(figsize=(10, 6))
sns.kdeplot(
    data=titanic_age,
    x="age",
    hue="survived",
    fill=True
)
plt.title("Age Distribution by Survival (KDE)")
plt.xlabel("Age")
plt.ylabel("Density")
plt.show()
```
